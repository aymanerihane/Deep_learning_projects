```python

# Load the pre-trained emotion detection model
model = load_model('model_CNN.h5')  # Update with your model path
labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=False, max_num_faces=5, min_detection_confidence=0.5)

# Open webcam feed
cap = cv2.VideoCapture(0)

while True:
    ret, video = cap.read()  # video is the frame

    # Check if frame was captured successfully
    if not ret:
        break

    # Flip the frame horizontally
    video = cv2.flip(video, 1)

    height, width, _ = video.shape

    # Convert image color from BGR to RGB
    rgb_video = cv2.cvtColor(video, cv2.COLOR_BGR2RGB)

    # Facial landmarks detection
    results = face_mesh.process(rgb_video)

    # Check if landmarks are detected
    if results.multi_face_landmarks:
        for facial_landmarks in results.multi_face_landmarks:
            x_min, y_min, x_max, y_max = width, height, 0, 0
            for landmark in facial_landmarks.landmark:
                x = int(landmark.x * width)
                y = int(landmark.y * height)
                
                # Update bounding box coordinates
                x_min = min(x_min, x)
                y_min = min(y_min, y)
                x_max = max(x_max, x)
                y_max = max(y_max, y)

            # Crop the face from the frame
            face_crop = video[y_min:y_max, x_min:x_max]
            if face_crop.size != 0:
                # Preprocess face for the model
                face_gray = cv2.cvtColor(face_crop, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
                face_resized = cv2.resize(face_gray, (48, 48)) / 255.0  # Resize and normalize
                face_reshaped = np.expand_dims(face_resized, axis=(0, -1))  # Reshape to (1, 48, 48, 1)

                # Predict expression
                predictions = model.predict(face_reshaped)
                expression = np.argmax(predictions)

                # Get the emotion label
                emotion = labels[expression]

                # Draw bounding box and emotion label
                cv2.rectangle(video, (x_min, y_min), (x_max, y_max), (255, 0, 0), 2)
                cv2.putText(video, emotion, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show the video feed
    cv2.imshow("Real-Time Facial Expression Recognition", video)

    # Exit loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
from tensorflow.keras.models import load_model

# Load pre-trained emotion detection model
model = load_model('fer2013_trained_model.h5')  # Update with your model path
labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

# Initialize MediaPipe Face Detection
mp_face_detection = mp.solutions.face_detection
face_detection = mp_face_detection.FaceDetection(min_detection_confidence=0.5)

# Open webcam
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.flip(frame, 1)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    results = face_detection.process(rgb_frame)

    if results.detections:
        for detection in results.detections:
            bboxC = detection.location_data.relative_bounding_box
            h, w, _ = frame.shape
            x, y, w, h = int(bboxC.xmin * w), int(bboxC.ymin * h), int(bboxC.width * w), int(bboxC.height * h)

            # Ensure bounding box is within frame limits
            x = max(0, x)
            y = max(0, y)
            w = min(frame.shape[1] - x, w)
            h = min(frame.shape[0] - y, h)

            # Crop the face
            face_crop = frame[y:y + h, x:x + w]
            if face_crop.size != 0:
                try:
                    # Convert to grayscale, resize, normalize
                    face_gray = cv2.cvtColor(face_crop, cv2.COLOR_BGR2GRAY)
                    face_resized = cv2.resize(face_gray, (48, 48)) / 255.0
                    face_reshaped = np.expand_dims(face_resized, axis=(0, -1))  # Reshape for model

                    # Predict emotion
                    predictions = model.predict(face_reshaped)
                    expression = np.argmax(predictions)
                    emotion = labels[expression]

                    # Draw bounding box and emotion label
                    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
                    cv2.putText(frame, emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                except Exception as e:
                    print(f"Error processing face: {e}")

    cv2.imshow("Real-Time Face Detection & Emotion Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
